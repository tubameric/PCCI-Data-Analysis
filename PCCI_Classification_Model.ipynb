```python
# Code snippet 1

# Importing the packages and the dataset
import pandas as pd
import numpy as np
data=pd.read_csv('https://raw.githubusercontent.com/nvamsimohan/DallasDSA/main/PCCI%20Cases%20Readmit30%20Data.csv')

# Displaying the data shape
data.shape
```

```python
# Code snippet 2
# Displaying the data columns
data.columns
```

```python
# Code snippet 3
# Importing missingno package to see the missing data proportion
import missingno as msno
msno.bar(data)
```

```python
# Code snippet 4
# Setting the target variable
y = 'Readmit30'
```

```python
# Code snippet 5
# Setting the ignored variables
ignored_cols = ['Admit_Week','Temperature','Insurance_Provider','Weight', 'Height','Pulse']
```

```python
# Code snippet 6
# Setting the numerical variables
num_cols = ['Age','Bmi','Pat_Pain_Score','ER_Visits', 'IP_Visits', 'Chronic_Conditions', 'Glucose','Cost_Of_Initial_Stay']
```

```python
# Code snippet 7
# Setting the categorical variables
cat_cols = ['Admit_Month', 'Gender', 'Marital_Status',
       'Tobacco_User', 'Depression', 'ICU', 'Drug_Abuse',
       'Mood_Disorder', 'Diabetes', 'Anxiety', 'Obesity', 'Dementia','Condition',
       'Care_Plan_Following_Discharge']
```

```python
# Code snippet 8
# Importing pycaret classification libraries
from pycaret.classification import *
```

```python
# Code snippet 9
# Configuring the ML experiement using setup function
classif_model = setup(data,
                      target=y,
                      ignore_features= ignored_cols,
                      categorical_features=cat_cols,
                      numeric_features=num_cols)
```

```python
# Code snippet 10
# Invoking the ML algortihms
compare_models()
```
